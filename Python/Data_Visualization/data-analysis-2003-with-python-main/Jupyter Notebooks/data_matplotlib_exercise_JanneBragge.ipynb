```python
# 4. Remove the symbols, the final two numbers from the 'Price' column and convert it to numbers
# 5. And plot it again
```

```python
## Task 17:
# Remove symbols and the final two numbers from the 'Price' column
prices = car_sales['Price'].str[1:-2]

#convert to num
price_num = [float(price.replace(',', '')) for price in prices]

# draw the graph
plt.figure(figsize=(6, 4), dpi=75)
plt.plot(price_num)

 
```

Next create another column which adds the total amount of sales and another one which shows what date the car was sold.

**Hint:** To add a column up cumulatively, look up the cumsum() function. And to create a column of dates, look up the date_range() function.

```python
# 6. Add a column called 'Total Sales' to car_sales which cumulatively adds the 'Price' column
# 7. Add a column called 'Sale Date' which lists a series of successive dates starting from today (your today)
# 8. View the car_sales DataFrame
```

```python
## Task 18:
import datetime

car_sales_df = pd.DataFrame(car_sales)

car_sales_df['Price'] = car_sales['Price'].str[1:-3]
car_sales_df['Price'] = [int(price.replace(',', '')) for price in car_sales_df['Price']]

# Add a column 'Total Sales' which cumulatively adds the 'Price' column
car_sales_df['Total Sales'] = car_sales_df['Price'].cumsum()
car_sales_df

# Add a column 'Sale Date' which lists a series of successive dates starting from today
today = datetime.date.today()
car_sales_df['Sale Date'] = pd.date_range(start=today, periods=len(car_sales_df))

car_sales_df

 
```

Now we've got a numeric column (`Total Sales`) and a dates column (`Sale Date`), let's visualize them.

```python
# 9. Use the plot() function to plot the 'Sale Date' column versus the 'Total Sales' column
```

```python
## Task 19:
import matplotlib.pyplot as plt

plt.figure(figsize=(6, 4), dpi=75)
plt.plot(car_sales_df['Sale Date'], car_sales_df['Total Sales'], linestyle='-')

 
```

```python
# 10. Convert the 'Price' column to the integers
# 11. Create a scatter plot of the 'Odometer (KM)' and 'Price' column using the plot() function
# 12. Name axes
```

```python
## Task 20:
# Convert 'Price' column to integers
car_sales_df['Price'] = car_sales_df['Price'].astype(int)

# Create scatter plot of 'Odometer (KM)' and 'Price'
plt.figure(figsize=(6, 4), dpi=75)
plt.scatter(car_sales_df['Odometer (KM)'], car_sales_df['Price'])

 
```

Completing this step correctly should result in a figure which looks like below.

```python
# Replicate the above plot in whichever way you see fit
# Note: There is different methods of doing it, yours might be different than mine. 
#       So your steps might differ steps below. 


# 1. Create DataFrame with patients over 50 years old
# 2. Create the plot
# 3. Plot the data (hint. use scatter parameter cmap = 'cool')
# 4. Plot the average "chol" line 
# 5. Customize the plot xlabel, ylabel and title
# 6. Add colorbar to fig
# 7. Add also legend as a reference for colorbar
```

```python
## Task 30:
# 1. Create DataFrame with patients over 50 years old
patients_over_50 = heart_disease[heart_disease['age'] > 50]

# 2. Create the plot
fig, ax = plt.subplots(figsize=(7, 5))

# 3. Plot the data
scatter = ax.scatter(patients_over_50['age'], patients_over_50['chol'], c=patients_over_50['trestbps'], cmap='cool', label='Data')

# 4. Plot the average "chol" line
average_chol = patients_over_50['chol'].mean()
ax.axhline(y=average_chol, color='green', linestyle='--', label='Average Cholesterol')

# 5. Customize the plot
ax.set_xlabel('Age', fontsize=12)
ax.set_ylabel('Cholesterol', fontsize=12)
ax.set_title('Heart Disease and Cholesterol Levels', fontsize=14)

# 6. Add colorbar to fig
cbar = fig.colorbar(scatter)
cbar.set_label('trestbps', fontsize=12)

# 7. Add legend as a reference for colorbar
legend_handles = []
legend_labels = []

# Add legend for trestbps levels
trestbps_levels = range(100, 201, 10)
for level in trestbps_levels:
    legend_handles.append(plt.Line2D([], [], color='grey', marker='o', linestyle='None', label=str(level)))
    legend_labels.append(str(level))

ax.legend(title='trestbps', handles=legend_handles, labels=legend_labels)

# Show plot
plt.show() 

 
```
