#### Step 3: Generate test data

```python
## Generate test data with following rows
# 1. time = np.arange(0, 2*np.pi, 0.1)
# 2. data = np.sin(t)
```

```python

time = np.arange(0, 2 * np.pi, 0.1)
data = np.sin(time)

# Print the generated data
print("Time:", time)
print("Data (sin):", data)

 
```

#### Step 4: Print shape of 'data' -variable a and plot it

```python
## 1. calculating length of the data with Numpy shape-command  
## 2. plotting a with pyplot plot(t, d)-command
```

```python
# Calculate the length of the data using Numpy's shape command
data_length = np.shape(data)[0]
print("Length of the data:", data_length)

# Plot the data using Matplotlib
plt.plot(time, data)
plt.xlabel('Time')
plt.ylabel('Sin(Time)')
plt.title('Sinusoidal Function')
plt.show()

```
